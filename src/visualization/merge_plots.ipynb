```python
from pathlib import Path

import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import matplotlib.image as mpimg

# Setup figure and grid
fig = plt.figure(figsize=(15, 15))
gs = gridspec.GridSpec(2, 1, height_ratios=[1, 1], hspace=0.01, wspace=-0.01)

base_path = Path("../../out/correlation")

images = [
    (base_path / 'corr_vs_linreg.png', gs[0, 0], 'A', 0.01),
    (base_path / 'corr_vs_train.png', gs[1, 0], 'B', 0.01),
]

for img_path, position, label, x_pos in images:
    ax = plt.subplot(position)
    ax.imshow(mpimg.imread(img_path))
    ax.axis('off')
    ax.text(x_pos, 0.975, label, fontsize=20, fontweight='bold',
            horizontalalignment='left', verticalalignment='top',
            transform=ax.transAxes)

plt.savefig(base_path / 'combined_figure.png', dpi=300, bbox_inches='tight')
plt.show()
```
